```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("netflix_titles.csv")
```

```python
country = (
    df['country']
    .dropna()
    .str.split(',')
    .explode()
)

country = country.str.strip()
```

```python
top_countries = (
    country
    .value_counts()
    .head(10)
)

top_countries
```

```python
plt.figure(figsize=(12,6))

top_countries.plot(kind='bar')

plt.title(
    'Top 10 Content Producing Countries'
)

plt.xlabel('Country')

plt.ylabel('Number of Titles')

plt.xticks(rotation=45)

plt.savefig(
    'country_analysis.png',
    bbox_inches='tight'
)

plt.show()
```

# Executive Insight: Country Expansion Analysis

## Business Question

Which countries contribute the most content to Netflix?

## Findings

- The United States remains the largest contributor.
- India is one of the strongest international content markets.
- The United Kingdom, Canada, and Japan maintain significant content volumes.

## Business Interpretation

Netflix's strategy demonstrates a balance between mature content markets and emerging international markets.

The strong representation of India highlights Netflix's commitment to regional content expansion.

## Recommendation

Continue investing in high-growth international markets while maintaining leadership in established regions.
