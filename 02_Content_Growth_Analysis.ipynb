```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("netflix_titles.csv")
```

```python
df['release_year'].describe()
```

```python
content_growth = (
    df['release_year']
    .value_counts()
    .sort_index()
)

content_growth.head()
```

```python
plt.figure(figsize=(12,6))

content_growth.plot()

plt.title("Netflix Content Growth Over Time")

plt.xlabel("Release Year")

plt.ylabel("Number of Titles")

plt.grid(True)

plt.show()
```

# Content Growth Analysis

## Business Question

How has Netflix expanded its content portfolio over time?

## Key Insight

The chart illustrates the growth trajectory of Netflix's content library and highlights periods of accelerated investment in content production and acquisition.

## Business Value

Understanding content growth trends helps evaluate Netflix's long-term content strategy and investment priorities.

# Executive Insight: Content Growth Analysis

## Business Question

How has Netflix's content portfolio evolved over time?

## Findings

- Content growth remained relatively low until the early 2000s.
- Significant expansion began after 2010.
- The strongest growth period occurred between 2015 and 2019.
- Content volume peaked around 2019.
- A decline is visible after 2020.

## Business Interpretation

Netflix aggressively increased content investment between 2015 and 2019 as part of its global expansion strategy.

The decline after 2020 may indicate:

- Increased focus on content quality over quantity
- Content production disruptions
- Strategic portfolio optimization

## Recommendation

Continue investing in high-performing content categories while optimizing production efficiency.
