# Data Quality Assessment

## Findings

The Netflix dataset contains several missing values that must be addressed before analysis.

### Major Issues

- Director information is missing for 2,634 titles (29.91%).
- Country information is missing for 831 titles (9.44%).
- Cast information is missing for 825 titles (9.37%).

### Minor Issues

- Date Added contains 10 missing records.
- Rating contains 4 missing records.

### Recommendation

Missing categorical values will be replaced with "Unknown" to preserve records while enabling analysis.**bold text**

## Duplicate Record Assessment

Duplicate records can distort reporting and business metrics.

The dataset was checked for duplicate entries before analysis.

```
df['director'] = df['director'].fillna('Unknown')

df['cast'] = df['cast'].fillna('Unknown')

df['country'] = df['country'].fillna('Unknown')

df['rating'] = df['rating'].fillna('Unknown')
```

```
df.isnull().sum()
```

```
df['date_added'] = df['date_added'].fillna('Not Available')
```
