```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("netflix_titles.csv")
```

```python
genre = (
    df['listed_in']
    .str.split(',')
    .explode()
)
```

```python
top_genres = genre.value_counts().head(10)

top_genres
```

```python
plt.figure(figsize=(12,6))

top_genres.plot(kind='bar')

plt.title(
    'Top 10 Netflix Genres'
)

plt.xlabel('Genre')

plt.ylabel('Number of Titles')

plt.xticks(rotation=45)
plt.savefig(
    "genre_strategy_Analysis.png",
    bbox_inches="tight"
)

plt.show()
```

# Executive Insight: Genre Strategy Analysis

## Business Question

Which content genres dominate Netflix's content portfolio?

## Key Findings

- International Movies represent the largest content category.
- Drama is the second largest genre segment.
- Comedies maintain a significant share of the content library.
- Documentary and Action content also contribute heavily to the platform.

## Business Interpretation

The strong presence of International Movies indicates Netflix's focus on global audience expansion.

Drama remains a core content pillar due to its broad audience appeal and consistent demand.

The diversity of genres demonstrates Netflix's strategy of serving multiple audience segments across regions.

## Strategic Recommendation

- Continue investing in International content.
- Expand successful Drama franchises.
- Increase localized content production in emerging markets.
- Evaluate growth opportunities in underrepresented genres.
