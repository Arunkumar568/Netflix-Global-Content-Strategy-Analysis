```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("netflix_titles.csv")
```

```python
df['rating'].value_counts()
```

```python
top_ratings = (
    df['rating']
    .value_counts()
    .head(10)
)

top_ratings
```

```python
plt.figure(figsize=(12,6))

top_ratings.plot(kind='bar')

plt.title(
    'Netflix Audience Rating Distribution'
)

plt.xlabel('Rating')

plt.ylabel('Number of Titles')

plt.xticks(rotation=45)

plt.savefig(
    'audience_analysis.png',
    bbox_inches='tight'
)

plt.show()
```

# Executive Insight: Audience Targeting Analysis

## Business Question

Which audience segments receive the most content investment?

## Findings

- TV-MA represents the largest content category.
- TV-14 also contributes a significant portion of the catalog.
- Family and children's content represent a smaller share.

## Business Interpretation

Netflix primarily targets mature and young-adult audiences.

The platform's content strategy appears focused on maximizing engagement among adult viewers while maintaining a balanced catalog for other audience groups.

## Strategic Recommendation

- Continue investing in high-performing mature audience content.
- Expand family-friendly content in emerging markets.
- Monitor audience preference shifts through rating-based analysis.
